```python
#Se carga primero el documento a Google colab
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
file_path = "/content/drive/MyDrive/novacredit_clientes.xlsx"
df_NovaCredits = pd.read_excel(file_path)
df_NovaCredits.head()
```

```python
#Identificamos valores Nulos
df_NovaCredits.isnull().sum()
```

```python
#Identificamos valores Nulos en porcentaje
(df_NovaCredits.isnull().sum() / len(df_NovaCredits)) * 100
```

```python
#Tratamiento de los valores nulos
df_NovaCredits['ingreso_mensual'].fillna(df_NovaCredits['ingreso_mensual'].mean(), inplace=True)

df_NovaCredits['metodo_pago'].fillna(df_NovaCredits['metodo_pago'].mode()[0], inplace=True)
```

```python
#Transformacion  y codificacion de variables categoricas
df_NovaCredits = pd.get_dummies(df_NovaCredits, columns=[
    'segmento_cliente',
    'region',
    'canal_adquisicion',
    'tipo_producto',
    'tipo_contrato',
    'metodo_pago'
], drop_first=True)
```

```python
#Identificacion de outliers
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
df_NovaCredits.boxplot(column=['ingreso_mensual'])
plt.show()
```

```python
#Tratamiento de outliers
Q1 = df_NovaCredits['ingreso_mensual'].quantile(0.25)
Q3 = df_NovaCredits['ingreso_mensual'].quantile(0.75)
IQR = Q3 - Q1

df_NovaCredits= df_NovaCredits[(df_NovaCredits['ingreso_mensual'] >= Q1 - 1.5*IQR) &
        (df_NovaCredits['ingreso_mensual'] <= Q3 + 1.5*IQR)]
```

#Decisiones tecnicas en la preparacion de tecnicas


### 1. Manejo de valores nulos
Se optó por imputar los valores faltantes en lugar de eliminarlos, con el objetivo de no perder información relevante:
- En variables numéricas (como ingreso_mensual) se utilizó la media, ya que representan datos continuos.
- En variables categóricas (metodo_pago) se utilizó la moda, al ser la categoría más frecuente.



### 2. Codificación de variables categóricas

Esto fue necesario porque:
- Los modelos de machine learning no pueden trabajar directamente con texto.
- Se evita introducir relaciones ordinales incorrectas entre categorías.



### 3. Identificación y tratamiento de outliers
Se empleó el método del rango intercuartílico (IQR) para detectar valores atípicos en variables numéricas.
